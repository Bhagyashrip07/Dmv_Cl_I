```python
# Convert the date/time column to datetime format (if not already)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
# Set the Size and Style
plt.figure(figsize=(12,6))
```

```python
# Plot the AQI data
plt.plot(df['Date'], df['AQI'], label='AQI', color='blue')
```

```python
# Plot overall trend
plt.figure(figsize=(12, 6))
plt.plot(df['AQI'], marker='o', linestyle='-')
plt.title('Overall AQI Trend Over Time')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.grid(True)
plt.show()
```

```python
# Plot PM2.5 trend
plt.figure(figsize=(12, 6))
plt.plot(df['PM2.5'], marker='o', linestyle='-')
plt.title('PM2.5 Trend Over Time')
plt.xlabel('Date')
plt.ylabel('PM2.5 Level')
plt.grid(True)
plt.show()
```

```python
# Plot PM10 trend
plt.figure(figsize=(12, 6))
plt.plot(df['PM10'], marker='o', linestyle='-')
plt.title('PM10 Trend Over Time')
plt.xlabel('Date')
plt.ylabel('PM10 Level')
plt.grid(True)
plt.show()
```

```python
# Plot CO trend
plt.figure(figsize=(12, 6))
plt.plot(df['CO'], marker='o', linestyle='-')
plt.title('CO Trend Over Time')
plt.xlabel('Date')
plt.ylabel('CO Level')
plt.grid(True)
plt.show()
```

```python
# Create box plot for AQI values
plt.figure(figsize=(12, 6))
plt.boxplot(data=df[['PM2.5', 'PM10', 'CO', 'AQI']])
plt.title('Distribution of AQI and Pollutant Levels')
plt.xlabel('Pollutants')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
